Here's a simple example of linear regression with code to help you understand how it works and how to plot the regression line.

## Import necessary libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Step 1: Generate synthetic but realistic data


```python
np.random.seed(42)  # For reproducibility

# Simulate 100 students: hours studied between 1 and 10
hours_studied = np.random.uniform(1, 10, size=100)

# Create test scores: positive correlation with noise added
# Base formula: score = 5 * hours + 30 + random noise
test_scores = 5 * hours_studied + 30 + np.random.normal(0, 5, size=100)

# Store in a DataFrame
df = pd.DataFrame({
    'HoursStudied': hours_studied,
    'TestScore': test_scores
})
df
```

```python
# Create a better-looking scatter plot
plt.figure(figsize=(8, 5))  # Set the figure size
plt.scatter(hours_studied, test_scores, color='teal', marker='o', s=50, alpha=0.7, edgecolor='black')

# Add labels and title
plt.title('Test Score vs Hours Studied', fontsize=14)
plt.xlabel('Hours Studied', fontsize=12)
plt.ylabel('Test Score', fontsize=12)

# Add grid for readability
plt.grid(True, linestyle='--', alpha=0.5)

# Show the plot
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

## Step 2: Fit linear regression model


```python
X = df[['HoursStudied']]  # Independent variable
y = df['TestScore']       # Dependent variable

model = LinearRegression()
model.fit(X, y)

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_

print(f"Regression Equation: y = {slope:.2f}x + {intercept:.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate synthetic data (100 students)
np.random.seed(42)
hours_studied = np.random.uniform(1, 10, size=100)
test_scores = 5 * hours_studied + 30 + np.random.normal(0, 5, size=100)

df = pd.DataFrame({
    'HoursStudied': hours_studied,
    'TestScore': test_scores
})

# Fit linear regression model
X = df[['HoursStudied']]  # Features (2D array needed for sklearn)
y = df['TestScore']       # Target

model = LinearRegression()
model.fit(X, y)

# Get regression line values
slope = model.coef_[0]
intercept = model.intercept_
x_values = np.array([df['HoursStudied'].min(), df['HoursStudied'].max()])
y_values = slope * x_values + intercept

# Plot with Matplotlib
plt.figure(figsize=(10, 6))
plt.scatter(df['HoursStudied'], df['TestScore'], color='red', alpha=0.6, label='Data Points')
plt.plot(x_values, y_values, color='blue', label='Regression Line')

# Add labels and title
plt.title('Test Score vs Hours Studied (Regression Line with Matplotlib)')
plt.xlabel('Hours Studied')
plt.ylabel('Test Score')

# Show legend and grid
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Multiple LR

```python
# Import necessary libraries

# NumPy for numerical operations (e.g., random number generation, array handling)
import numpy as np

# Pandas for data manipulation and creating DataFrames
import pandas as pd

# LinearRegression from sklearn.linear_model to build the regression model
from sklearn.linear_model import LinearRegression

# Set a random seed for reproducibility
# This ensures that the same random numbers are generated every time the code runs
np.random.seed(42)

# Define the number of samples (i.e., number of students in this case)
n_samples = 100

# Generate synthetic data for 'Hours Studied'
# Random float values between 1 and 10 for 100 students
hours_studied = np.random.uniform(1, 10, size=n_samples)

# Generate synthetic data for 'Sleep Hours'
# Random float values between 4 and 10 hours of sleep
sleep_hours = np.random.uniform(4, 10, size=n_samples)

# Generate synthetic test scores based on:
# - Positive effect of hours studied (+5 per hour)
# - Negative effect of sleep hours (-2 per hour)
# - Some random noise (normal distribution around 0 with std deviation of 3)
test_scores = 5 * hours_studied - 2 * sleep_hours + np.random.normal(0, 3, size=n_samples)

# Create a Pandas DataFrame to organize the data into columns
df = pd.DataFrame({
    'HoursStudied': hours_studied,   # First input feature
    'SleepHours': sleep_hours,       # Second input feature
    'TestScore': test_scores         # Target/output variable
})
```

```python
# Define input (X) and output (y)

# X = input features (also called predictors or independent variables)
# We're using two columns: 'HoursStudied' and 'SleepHours'
X = df[['HoursStudied', 'SleepHours']]

# y = output/target variable (also called dependent variable)
# This is what we're trying to predict: 'TestScore'
y = df['TestScore']


# Create and fit the linear regression model

# Create a Linear Regression model object
model = LinearRegression()

# Fit the model to the data
# This finds the best-fit line (or plane, in this case) that minimizes prediction error
model.fit(X, y)


# Get coefficients and intercept

# The coefficients represent how much each feature influences the target
coefficients = model.coef_

# The intercept is the value of y when all features are 0
intercept = model.intercept_


# Print out the regression equation so we can interpret the model

print(f"Equation: TestScore = {coefficients[0]:.2f} * HoursStudied + {coefficients[1]:.2f} * SleepHours + {intercept:.2f}")
print(f"Equation: y = {coefficients[0]:.2f} * X-1 + {coefficients[1]:.2f} * X-2 + {intercept:.2f}")
```

```python
# Import necessary libraries for plotting in 3D
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Enables 3D plotting

# Create a new figure with specified size (width=10 inches, height=7 inches)
fig = plt.figure(figsize=(10, 7))

# Add a 3D subplot to the figure
# 'projection="3d"' enables 3D axes
ax = fig.add_subplot(111, projection='3d')

# Plot the actual data points in 3D space
# x-axis: Hours Studied
# y-axis: Sleep Hours
# z-axis: Test Score
ax.scatter(df['HoursStudied'], df['SleepHours'], df['TestScore'],
           color='blue', label='Data Points')

# Generate a grid of x and y values for the regression plane
# linspace(): creates evenly spaced values between min and max of each feature
x_surf, y_surf = np.meshgrid(np.linspace(df['HoursStudied'].min(), df['HoursStudied'].max(), 20),
                             np.linspace(df['SleepHours'].min(), df['SleepHours'].max(), 20))

# Flatten the 2D grids into 1D arrays so we can pass them to model.predict()
x_surf_flat = x_surf.flatten()
y_surf_flat = y_surf.flatten()

# Stack the flattened arrays into a 2D array of shape (N, 2), suitable for prediction
xy = np.vstack([x_surf_flat, y_surf_flat]).T

# Use the trained linear regression model to predict z-values (test scores) for all (x, y) pairs
z_surf = model.predict(xy).reshape(x_surf.shape)  # Reshape back to the original 2D grid shape

# Plot the regression plane on top of the scatter plot
ax.plot_surface(x_surf, y_surf, z_surf,
                color='red', alpha=0.5,  # Set transparency
                rstride=1, cstride=1,    # Controls resolution of the surface
                label='Regression Plane')

# Add axis labels for clarity
ax.set_xlabel('Hours Studied')
ax.set_ylabel('Sleep Hours')
ax.set_zlabel('Test Score')
ax.view_init(elev=10, azim=100)
# Add a title to the plot
plt.title('3D Regression Plane')

# Show legend to differentiate between data points and regression plane
plt.legend()

# Display the final plot
plt.show()
```

```python
# Use the trained model to make predictions on the input data (X)
y_pred = model.predict(X)

# Calculate residuals: difference between actual values (y) and predicted values (y_pred)
# Residuals = y - y_pred
residuals = y - y_pred

# Create a new figure for the residual plot
plt.figure(figsize=(8, 5))

# Plot residuals vs predicted values
# This helps us check if our model has any patterns in its errors
plt.scatter(y_pred, residuals, alpha=0.7, color='blue', edgecolor='k')

# Add a horizontal red line at y=0 to indicate perfect predictions
plt.axhline(y=0, color='r', linestyle='-')

# Add title and axis labels for clarity
plt.title('Residual Plot')
plt.xlabel('Predicted Scores')      # x-axis: predicted values
plt.ylabel('Residuals')             # y-axis: error (actual - predicted)

# Add gridlines for better readability
plt.grid(True)

# Adjust layout so everything fits nicely
plt.tight_layout()

# Display the plot
plt.show()
```
